```python
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
```

```python
texts = ["wiper", "wiper blade", "windshield wiper", "windshield wiper blade", "door handle", "door lock", "door lock mechanism", "door lock mechanism assembly", "door lock mechanism assembly assembly", "door lock mechanism assembly assembly assembly"]

embeddings = model.encode(texts)

# perform dbscan clustering
dbscan = DBSCAN(eps=0.2, min_samples=2, metric='cosine')
dbscan.fit(embeddings)


print(dbscan.labels_)

for i, label in enumerate(dbscan.labels_):
    print(f"{texts[i]} -> {label}")
```
